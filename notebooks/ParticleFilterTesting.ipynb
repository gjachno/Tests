```python
mu_TC = 0
sigma_TC = 0.25
mu_Y = 100
sigma_Y = 0.5

def f_X(x):
    return sp.stats.lognorm(sigma_TC,loc=mu_TC).pdf(x)

def f_Z1gY(z, y):
    return f_X(z-y)

def f_Z2gY(z, y):
    return f_X(y-z)

def f_Y(y):
    return sp.stats.norm(mu_Y, sigma_Y).pdf(y)

N = NormalDistr(mu_TC,sigma_TC)
X = exp(N)
Y = NormalDistr(mu_Y,sigma_Y)
Z1 = X + Y
Z2 = Y - X

def f_Z1(z):
    return Z1.pdf(z)

def f_Z2(z):
    return Z2.pdf(z)

def f_YgZ1(y, z):
    return f_Z1gY(z,y)*f_Y(y)/f_Z1(z)

def f_YgZ2(y, z):
    return f_Z2gY(z,y)*f_Y(y)/f_Z2(z)
```

```python
def getY1(mu_X, sigma_X, mu_TC, sigma_TC):
    X = NormalDistr(mu_X,sigma_X)
    N = NormalDistr(mu_TC,sigma_TC)
    TC = exp(N)
    Y1 = X + TC
    return Y1

def getY2(mu_X, sigma_X, mu_TC, sigma_TC):
    X = NormalDistr(mu_X,sigma_X)
    N = NormalDistr(mu_TC,sigma_TC)
    TC = exp(N)
    Y2 = X - TC
    return Y2

def f_XgY1(x, y, mu_X, sigma_X, mu_TC, sigma_TC, f_Y1):
    f_X = sp.stats.norm(mu_X,sigma_X).pdf(x)
    f_Y1gX = sp.stats.lognorm(sigma_TCt,loc=mu_TCt).pdf(y-x)
    return f_Y1gX*f_X/f_Y1

def f_XgY2(x, y, mu_X, sigma_X, mu_TC, sigma_TC, f_Y2):
    f_X = sp.stats.norm(mu_X,sigma_X).pdf(x)
    f_Y2gX = sp.stats.lognorm(sigma_TCt,loc=mu_TCt).pdf(x-y)
    return f_Y2gX*f_X/f_Y2

def sample_XgY1(y, mu_X, sigma_X, mu_TC, sigma_TC, f_Y1, N):
    # find max of pdf
    x_grid = np.arange(mu_X-5*sigma_X,y,0.1)
    f_XgY1_graph = np.ndarray((len(x_grid),1))
    for i in range(0,len(x_grid)):
        f_XgY1_graph[i] = f_XgY1(x_grid[i], y, mu_X, sigma_X, mu_TC, sigma_TC, f_Y1)
    # print(str(len(f_XgY1_graph)))
    max_pdf = np.max(f_XgY1_graph)
    sample = np.ndarray((N,1))
    count = 0
    while (count<N):
        x_prop = np.random.uniform(mu_X-5*sigma_X,y,1)[0]
        test_point = np.random.uniform(0,max_pdf,1)
        if (f_XgY1(x_prop, y, mu_X, sigma_X, mu_TC, sigma_TC, f_Y1) > test_point):
            sample[count] = x_prop
            count = count+1
            #if (count % 100 == 0):
                #print("Count = " + str(count))
        else:
            pass

    return sample

def par_sample_XgY1(y, mu_X, sigma_X, mu_TC, sigma_TC, N):
    num_cores = min(multiprocessing.cpu_count(),10)
    Y1 = getY1(mu_X, sigma_X, mu_TC, sigma_TC)
    f_Y1 = Y1.pdf(y)
    samples = Parallel(n_jobs=num_cores)(
      delayed(sample_XgY1)(y, mu_X, sigma_X, mu_TC, sigma_TC, f_Y1, N/num_cores) for i in range(num_cores)
    )    
    return np.concatenate(samples)

def sample_XgY2(y, mu_X, sigma_X, mu_TC, sigma_TC, f_Y2, N):
    # find max of pdf
    x_grid = np.arange(y,mu_X+5*sigma_X,0.1)
    f_XgY2_graph = np.ndarray((len(x_grid),1))
    for i in range(0,len(x_grid)):
        f_XgY2_graph[i] = f_XgY2(x_grid[i], y, mu_X, sigma_X, mu_TC, sigma_TC, f_Y2)
    max_pdf = np.max(f_XgY2_graph)
    sample = np.ndarray((N,1))
    count = 0
    while (count<N):
        x_prop = np.random.uniform(y,mu_X+5*sigma_X,1)[0]
        test_point = np.random.uniform(0,max_pdf,1)
        if (f_XgY2(x_prop, y, mu_X, sigma_X, mu_TC, sigma_TC, f_Y2) > test_point):
            sample[count] = x_prop
            count = count+1
            #if (count % 100 == 0):
                #print("Count = " + str(count))
        else:
            pass

    return sample

def par_sample_XgY2(y, mu_X, sigma_X, mu_TC, sigma_TC, N):
    num_cores = min(multiprocessing.cpu_count(),10)
    Y2 = getY2(mu_X, sigma_X, mu_TC, sigma_TC)
    f_Y2 = Y2.pdf(y)
    samples = Parallel(n_jobs=num_cores)(
      delayed(sample_XgY2)(y, mu_X, sigma_X, mu_TC, sigma_TC, f_Y2, N/num_cores) for i in range(num_cores)
    )    
    return np.concatenate(samples)

def multinomial_flatten(parent, counts):
    N = np.sum(counts)
    children = np.full((1,N), np.nan)
    index = 0
    for i in range(0,len(counts)):
        if (counts[i]>0):
            for j in range(0,counts[i]):
                # print("Adding " + str(parent[i]))
                children[0][index] = deepcopy(parent[i])
                index = index + 1
    return children
```
